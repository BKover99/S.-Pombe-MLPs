# In this Jupyter notebook I will generate 96-well layouts as individual csv files from the list of deletion strains in the deletion library

```python
import pandas as pd
import numpy as np
```

```python
#import /Users/bencekover/Library/CloudStorage/OneDrive-UniversityCollegeLondon/MSci Bahler lab/S.-Pombe-biofilm/external data/Bioneer_collection_version_5_Prototroph.xlsx
bioneer = pd.read_excel('/Users/bencekover/Library/CloudStorage/OneDrive-UniversityCollegeLondon/MSci Bahler lab/S.-Pombe-biofilm/external data/bioneer/Bioneer_collection_version_5_Prototroph.xlsx', sheet_name='Sheet1')
bioneer = bioneer.append([pd.Series([np.nan]*len(bioneer.columns), index=bioneer.columns)]*192, ignore_index=True)
```

```python
#iterate through the rows of the dataframe in 96 well format
n_plates = len(bioneer)/96
#generate a list of red, blue, yellow, green repeated 10 times
colours = ["red", "blue", "yellow", "green"]*10
plates = [i for i in range(1,11) for j in range(0,4)]
for i in range(0, int(n_plates)):
      temp_df = bioneer.iloc[i*96:(i+1)*96]
      #turn this into a 96 well format. THe first 12 values of SystematicID are the first row of the plate, the next 12 are the second row etc. And then save the resulting dataframe as a csv file
      #create empty df with 12 columns and 8 rows
      temp_df_96 = pd.DataFrame(index=range(0,8), columns=range(0,12))
      #iterate through the rows of the dataframe
      for j in range(0,8):
            #iterate through the columns of the dataframe and extraact the systematic ID values
            for k in range(0,12):
                  temp_df_96.iloc[j,k] =  temp_df.iloc[j*12+k]["SystematicID"]
      #save the dataframe as a csv file
      temp_df_96.to_csv("/Users/bencekover/Library/CloudStorage/OneDrive-UniversityCollegeLondon/MSci Bahler lab/S.-Pombe-biofilm/external data/bioneer/96_well_layout_plate_"+str(plates[i])+"_"+colours[i]+".csv", index=False, header=False)
```

```python
#add 184 nan rows and create 10 384 plates for bioneer and save them with 384 in name
#add 184 nan rows

n_plates = len(bioneer)/384
for i in range(1, int(n_plates+1)):
      #load red file for the plate
      red = pd.read_csv("/Users/bencekover/Library/CloudStorage/OneDrive-UniversityCollegeLondon/MSci Bahler lab/S.-Pombe-biofilm/external data/bioneer/96_well_layout_plate_"+str(i)+"_red.csv", header=None)
      blue = pd.read_csv("/Users/bencekover/Library/CloudStorage/OneDrive-UniversityCollegeLondon/MSci Bahler lab/S.-Pombe-biofilm/external data/bioneer/96_well_layout_plate_"+str(i)+"_blue.csv", header=None)
      yellow = pd.read_csv("/Users/bencekover/Library/CloudStorage/OneDrive-UniversityCollegeLondon/MSci Bahler lab/S.-Pombe-biofilm/external data/bioneer/96_well_layout_plate_"+str(i)+"_yellow.csv", header=None)
      green = pd.read_csv("/Users/bencekover/Library/CloudStorage/OneDrive-UniversityCollegeLondon/MSci Bahler lab/S.-Pombe-biofilm/external data/bioneer/96_well_layout_plate_"+str(i)+"_green.csv", header=None)
      #construct the 384 as 1 from red, 1 from blue 12 times, then 1 from yellow, 1 from green 12 times and so on
      temp_df_384 = pd.DataFrame(index=range(0,16), columns=range(0,24))
      
      for j in range(0,16):
            if j%2 == 0:
                  for k in range(0,24):
                        if k%2 == 0:
                              temp_df_384.iloc[j,k] = red.iloc[j//2,k//2]
                        else:
                              temp_df_384.iloc[j,k] = blue.iloc[j//2,k//2]
            elif j%2 == 1:
                  for k in range(0,24):
                        if k%2 == 0:
                              temp_df_384.iloc[j,k] = yellow.iloc[j//2,k//2]
                        else:
                              temp_df_384.iloc[j,k] = green.iloc[j//2,k//2]
      #save the dataframe as a csv file
      temp_df_384.to_csv("/Users/bencekover/Library/CloudStorage/OneDrive-UniversityCollegeLondon/MSci Bahler lab/S.-Pombe-biofilm/external data/bioneer/384_well_layout_plate_"+str(i)+".csv", index=False, header=False)
                        
```

# Here I will also generate 96 layouts for the ncRNA deletion library, based on the same logic.

```python
ncrna = "/Users/bencekover/Library/CloudStorage/OneDrive-UniversityCollegeLondon/MSci Bahler lab/S.-Pombe-biofilm/external data/Rodriguez-Lopez et al. ncRNA deletion library/ncRNA_deletions_lists.xlsx"
ncrna_df = pd.read_excel(ncrna, header = None)

ncrna_df_1 = ncrna_df.iloc[0:16]
ncrna_df_2 = ncrna_df.iloc[20:]
ncrna_df_2.reset_index(drop=True, inplace=True)
```

```python
#generate a list of red, blue, yellow, green
colours = ["red", "blue", "yellow", "green"]
for k in colours:
    #create a temporary dataframe
      temp_df = pd.DataFrame(index=range(0,8), columns=range(0,12))

      for i in range(0,12):
            for j in range(0,8):
                  if k == "red":
                        temp_df.iloc[j,i] = ncrna_df_1.iloc[j*2,i*2]
                        
                  elif k == "blue":
                        temp_df.iloc[j,i] = ncrna_df_1.iloc[j*2,i*2+1]
                  elif k == "yellow":
                        temp_df.iloc[j,i] = ncrna_df_1.iloc[j*2+1,i*2]
                  elif k== "green":
                        temp_df.iloc[j,i] = ncrna_df_1.iloc[j*2+1,i*2+1]
      #print(temp_df)

      temp_df.to_csv("/Users/bencekover/Library/CloudStorage/OneDrive-UniversityCollegeLondon/MSci Bahler lab/S.-Pombe-biofilm/external data/Rodriguez-Lopez et al. ncRNA deletion library/{}_{}_plate_{}.csv".format("ncRNA", k, 1), index=False)



        

```

```python
#repeat for the second plate
colours = ["red", "blue", "yellow", "green"]
for k in colours:
    #create a temporary dataframe
      temp_df = pd.DataFrame(index=range(0,8), columns=range(0,12))

      for i in range(0,12):
            for j in range(0,8):
                  if k == "red":
                        temp_df.iloc[j,i] = ncrna_df_2.iloc[j*2,i*2]
                        
                  elif k == "blue":
                        temp_df.iloc[j,i] = ncrna_df_2.iloc[j*2,i*2+1]
                  elif k == "yellow":
                        temp_df.iloc[j,i] = ncrna_df_2.iloc[j*2+1,i*2]
                  elif k== "green":
                        temp_df.iloc[j,i] = ncrna_df_2.iloc[j*2+1,i*2+1]
      #print(temp_df)

      temp_df.to_csv("/Users/bencekover/Library/CloudStorage/OneDrive-UniversityCollegeLondon/MSci Bahler lab/S.-Pombe-biofilm/external data/Rodriguez-Lopez et al. ncRNA deletion library/{}_{}_plate_{}.csv".format("ncRNA", k, 2), index=False)

    
```
